```python
# make node names snake_case
import re
import pandas as pd
```

```python
def to_snake(s: str) -> str:
    s = str(s).strip().lower()
    # replace any run of non-alphanumeric/underscore with a single underscore
    s = re.sub(r"[^\w]+", "_", s)
    # collapse repeated underscores and trim edges
    s = re.sub(r"_+", "_", s).strip("_")
    return s
```

```python
nodes = pd.read_csv('data/nodes.csv', sep= ',')
nodes
```

```python
nodes_snake = nodes.copy()
nodes_snake['node_name'] = nodes['node_name'].astype(str).map(to_snake)
```
